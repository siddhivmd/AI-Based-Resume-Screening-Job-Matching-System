```python
import pandas as pd
import numpy as np
import re
import nltk

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

nltk.download('stopwords')
from nltk.corpus import stopwords
df = pd.read_csv("/Resume.csv")  # change filename

print(df.head())
print(df.columns)
stop_words = set(stopwords.words('english'))

def clean_text(text):
    text = text.lower()
    text = re.sub(r'[^a-zA-Z ]', '', text)
    words = text.split()
    words = [word for word in words if word not in stop_words]
    return " ".join(words)

df['Resume_html'] = df['Resume_html'].apply(clean_text)
df['Category'] = df['Category'].apply(clean_text)
vectorizer = TfidfVectorizer(max_features=5000)

resume_vectors = vectorizer.fit_transform(df['Resume_html'])
job_vectors = vectorizer.transform(df['Category'])
similarity_scores = cosine_similarity(resume_vectors, job_vectors)

df['match_score'] = similarity_scores.diagonal() * 100
df_sorted = df.sort_values(by='match_score', ascending=False)

print(df_sorted[['match_score']].head(10))
def match_resume(resume_text, job_description):
    resume_text = clean_text(resume_text)
    job_description = clean_text(job_description)

    vectors = vectorizer.transform([resume_text, job_description])
    score = cosine_similarity(vectors[0], vectors[1])[0][0]

    return round(score * 100, 2)

```
